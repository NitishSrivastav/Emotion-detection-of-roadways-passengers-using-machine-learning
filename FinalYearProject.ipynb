**Data Imputation and Manipulation**

```python
#define a person impute median

def impute_median(series):
  return series.fillna(series.median())
```

```python
df.Rating = df['Rating'].transform(impute_median)
```

```python
df.isnull().sum()
```

```python
#modes of categorical values

print(df['Store'].mode())
print(df['Subject'].mode())
print(df['App ID'].mode())
print(df['Review ID'].mode())
print(df['Country'].mode())
print(df['Version'].mode())
print(df['Date'].mode())
print(df['Author'].mode())
print(df['Body'].mode())
print(df['Translated Subject'].mode())
print(df['Translated Body'].mode())
print(df['Sentiment'].mode())
print(df['Device'].mode())
print(df['Language'].mode())
print(df['OS Version'].mode())
print(df['Reply URL'].mode())
print(df['Topics'].mode())
print(df['Custom Topics'].mode())
print(df['Tags'].mode())
```

```python
df['Store'].fillna(str(df['Store'].mode().values[0]), inplace = True)
df['Subject'].fillna(str(df['Subject'].mode().values[0]), inplace = True)
df['App ID'].fillna(str(df['App ID'].mode().values[0]), inplace = True)
df['Review ID'].fillna(str(df['Review ID'].mode().values[0]), inplace = True)
df['Country'].fillna(str(df['Country'].mode().values[0]), inplace = True)
df['Version'].fillna(str(df['Version'].mode().values[0]), inplace = True)
df['Date'].fillna(str(df['Date'].mode().values[0]), inplace = True)
df['Author'].fillna(str(df['Author'].mode().values[0]), inplace = True)
df['Body'].fillna(str(df['Body'].mode().values[0]), inplace = True)
#df['Translated Subject'].fillna(str(df['Translated Subject'].mode().values[0]), inplace = True)
##df['Translated Body'].fillna(str(df['Translated Body'].mode().values[0]), inplace = True)
df['Sentiment'].fillna(str(df['Sentiment'].mode().values[0]), inplace = True)
#df['Device'].fillna(str(df['Device'].mode().values[0]), inplace = True)
df['Language'].fillna(str(df['Language'].mode().values[0]), inplace = True)
#df['OS Version'].fillna(str(df['OS Version'].mode().values[0]), inplace = True)
df['Reply URL'].fillna(str(df['Reply URL'].mode().values[0]), inplace = True)
df['Topics'].fillna(str(df['Topics'].mode().values[0]), inplace = True)
df['Custom Topics'].fillna(str(df['Custom Topics'].mode().values[0]), inplace = True)
df['Tags'].fillna(str(df['Tags'].mode().values[0]), inplace = True)
```
